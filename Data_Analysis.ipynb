```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your dataset
data = pd.read_excel('sanjuan_dataset.xlsx')

# Create DataFrame
df = pd.DataFrame(data)
```

```python
#10 most reported incident type in San Juan
most_reported_crime = data['incident_type'].value_counts().head(10)
print(most_reported_crime)
```

```python
#10 least reported incident type in San Juan
least_reported_crime = data['incident_type'].value_counts().tail(10)
print(least_reported_crime)
```

```python
#Number of crimes on each barangay
highest_incident = data['barangay'].value_counts().head(21)
print(highest_incident)
```

```python
#Types of crime in San Juan City
types_crime = data['incident_type'].unique()
print(types_crime)
```

```python
#Crimes per month from 2019 to 2023

# Convert 'date' column to datetime format
df['date_committed'] = pd.to_datetime(df['date_committed'])

# Extract month and year from the date
df['month'] = df['date_committed'].dt.to_period('M')

# Calculate the average number of crimes per month
average_crimes_per_month = df.groupby('month').size().reset_index(name='total_crimes')
average_crimes_per_month['average_crimes'] = average_crimes_per_month['total_crimes'].mean()

# Display the result
print(average_crimes_per_month[['month', 'total_crimes']])
```

```python
#Average Crime Monthly
average_crime = average_crimes_per_month['average_crimes'].unique()
print("Average Monthly Crime in San Juan: ", average_crime)
```

```python
barangay_names = data['barangay'].unique()
print(barangay_names)
```

```python
# Calculate total crimes per barangay
total_crimes_per_barangay = df['barangay'].value_counts().reset_index()
total_crimes_per_barangay.columns = ['barangay', 'total_crimes']

# Display the result
print(total_crimes_per_barangay)
```

```python
# Calculate total crimes per month and barangay
total_crimes = df.groupby(['month', 'barangay']).size().reset_index(name='total_crimes')

# Bar Chart: Total Crimes per Barangay
plt.figure(figsize=(12, 6))
sns.barplot(data=total_crimes, x='month', y='total_crimes', hue='barangay')
plt.title('Total Crimes per Barangay per Month')
plt.xlabel('Month')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45)
plt.legend(title='Barangay')
plt.tight_layout()
plt.show()

# Line Graph: Crime Trends Over Time
plt.figure(figsize=(12, 6))
#sns.lineplot(data=total_crimes, x='month', y='total_crimes', hue='barangay', marker='o')
plt.title('Crime Trends Over Time by Barangay')
plt.xlabel('Month')
plt.ylabel('Total Crimes')
plt.xticks(rotation=45)
plt.legend(title='Barangay')
plt.tight_layout()
plt.show()
```
